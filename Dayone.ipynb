```python
import pandas as pd
```

```python
df = pd.read_csv('Covid Data.csv')
```

```python
df
```

SELECTING MALE PATIENTS

```python
df = df[df['SEX']!=1]
df
```

```python
#Adding new column indicating those who died and  those who are alive in the covid data set
df['STATUS'] =df['DATE_DIED'].apply(lambda x : 'Cured' if '9999-99-99' in x.lower() else 'Died') 
df
```

```python
#Showing the day and month a patient died
df['DAY_DIED']= df['DATE_DIED'].apply(lambda x : x.split('/')[0])
df['MONTH_DIED']= df['DATE_DIED'].apply(lambda x : x.split('/')[1:-1])#printing middle value of list with three members
df.sample(50)
```

```python
#Number of people who died in 2020 and those that healed
df.STATUS.value_counts()
```

```python
#returning columns
df.columns
```

```python
df_out = df.drop(['MEDICAL_UNIT','USMER','SEX','PATIENT_TYPE','INMSUPR','HIPERTENSION','RENAL_CHRONIC','HIPERTENSION','CLASIFFICATION_FINAL','PREGNANT', 'COPD',],axis =1) 
```
