```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('My Males Covid Data.csv')
```

```python
df
```

```python
df['STATUS'] = df['DATE_DIED'].apply(lambda x : 1 if '9999-99-99' in x else 0)
df
```

```python
df.columns
```

```python
df['DAY_DIED'] = df['DAY_DIED'].apply(lambda x : 1 if '9999-99-99' in x else x)
```

```python
df.columns
```

```python
df_model =df[['AGE','INTUBED','DIABETES','ICU','CARDIOVASCULAR','ASTHMA','STATUS','PNEUMONIA','DAY_DIED']]
df_model
```

```python
#Removing unknown values
for x in df_model.columns:
    df_model = df_model[df_model[x]!=97]
    df_model = df_model[df_model[x]!=98]
```

```python
df_model
```

```python
df_dummy = pd.get_dummies(df_model)
```

```python
df_dummy
```

```python
from sklearn.model_selection import train_test_split as ttt
x = df_dummy.drop('AGE', axis = 1)
y = df_dummy.AGE.values
X_train,X_test,y_train,y_test = ttt(x,y ,test_size = 0.2, random_state = 42)
```

```python
X_train
```

```python
y_train
```

MODEL BUILDING LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lr = LinearRegression()
scores = cross_val_score(lr,X_train,y_train, scoring = 'neg_mean_absolute_error')
scores
lr.fit(X_train,y_train)
```

```python
np.mean(scores)
```

```python
from sklearn.linear_model import Lasso
lr_l = Lasso()
lr_l.fit(X_train,y_train)
scores = cross_val_score(lr_l,X_train,y_train, scoring = 'neg_mean_absolute_error')
scores
```

```python
np.mean(scores)
```

MODEL BUILDING LASO REGRESSION

```python
from sklearn.ensemble import RandomForestRegressor
lf_r = RandomForestRegressor()
lf_r.fit(X_train,y_train)
scores = cross_val_score(lf_r,X_train,y_train, scoring = 'neg_mean_absolute_error')
scores
```

```python
np.mean(scores)
```

```python
tpred_lr =lr.predict(X_test)
tpred_lr_l =lr_l.predict(X_test)
tpred_lf_r =lf_r.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error as mae
mae(y_test,tpred_lr)
```

```python
mae(y_test,tpred_lr_l)
```

```python
mae(y_test,tpred_lf_r)
```

PRODUCTION

```python
#saving model using pickle
import pickle
pick1 = {'model': lf_r}
pickle.dump(pick1,open('model_f'+".p","wb"))
```

```python
file_name = "model_f.p"
with open(file_name ,'rb')as pickled:
    data = pickle.load(pickled)
    model = data['model']
model.predict(X_test.iloc[1,:].values.reshape(1,-1))
```
